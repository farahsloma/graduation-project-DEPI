# Feature Engineering 

#### 1 - total ridership :  total ridership for each day by summing all transport

```python
df["Total_ridership"] = df["Buses: Total Estimated Ridership"] + df["LIRR: Total Estimated Ridership"] + df["Metro-North: Total Estimated Ridership"]+df["Access-A-Ride: Total Scheduled Trips"] + df["Staten Island Railway: Total Estimated Ridership"]+ df["Subways: Total Estimated Ridership"]
```

#### 2 - Ridership Category : Categorizing ridership levels based on thresholds 

```python
def category_fun(x):
    if x > 1000000 and x < 3000000 :
        return "Moderate"
    elif x >= 3000000 and x < 5000000 :
        return "High"
    elif x > 5000000 :
        return "Very High"
    else : 
        return "Low"
    
```

```python
df["category"] = df["Total_ridership"].apply(category_fun)
```

####  3 -weighted_recovery : 
It calculates a weighted sum of recovery percentages for 
each mode, adjusted by their respective ridership volumes, and divides this sum by the 
total ridership across all modes. This metric provides a holistic view of the system's 
recovery status, ensuring that no single mode disproportionately influences the overall 
recovery percentage 

```python
df["weighted_recovery"] = df['Subways: % of Comparable Pre-Pandemic Day'] + df['Buses: % of Comparable Pre-Pandemic Day'] + df['LIRR: % of Comparable Pre-Pandemic Day'] + df['Metro-North: % of Comparable Pre-Pandemic Day'] + df['Access-A-Ride: % of Comparable Pre-Pandemic Day'] +df['Bridges and Tunnels: % of Comparable Pre-Pandemic Day'] + df["Staten Island Railway: % of Comparable Pre-Pandemic Day"] 
```

```python
df["weighted_recovery"] = df["weighted_recovery"] / 7
```

####  4 -  Weighted_recovery_category
Classifying recovery levels based on weighted recovery percentage:

```python
def category_wighted_recovery(x) :
    if x >= 100 :
        return "Fully Recovery"
    elif x >= 80 :
        return "High Recovery"
    elif x >= 50 :
        return "Medium Recovery"
    else :
        return "Low Recovery"
```

```python
df["Weighted_recovery_category"] = df["weighted_recovery"].apply(category_wighted_recovery)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

#### 5 - calender table
    The CalendarTable generates a structured date table that helps in time-based analysis. It includes 
    additional attributes for each date to enable easier grouping and filtering in reports. 

```python
df['weekend'] = df['Date'].dt.weekday.isin([5 ,6])
```

```python
df.head()
```

```python
def season (x):
    if x == 12 or x == 1 or x == 2  :
        return "winter"
    elif x == 4 or x == 3 or x == 5 :
        return "spring"
    elif x == 6 or x == 7 or x == 8 :
        return "summer"
    else :
        return "fall"
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Month_Name"] = df["Date"].dt.month_name()
df["Day"] = df["Date"].dt.day
df["Weekday"] = df["Date"].dt.weekday  
df["Weekday_Name"] = df["Date"].dt.day_name()

df["Quarter"] = df["Date"].dt.quarter
```

```python
df["season"] = df["Month"].apply(season)
```
